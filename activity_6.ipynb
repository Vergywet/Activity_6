```python
wins = football['Wins']
losses = football['Losses']

plt.figure(figsize=(5,5))
plt.title('Bar Graph for Wins vs  Losses')
plt.bar(wins,losses,color="blue",alpha=0.6)
plt.xlabel('Wins')
plt.ylabel('Losses')
plt.show()
```

```python

football =  np.random.normal(loc=20.0,scale=5.0,size = 365)
plt.hist(football,bins=20,color = "Purple",alpha= 0.7)
plt.title('Histograph of  Goals and  Assist')
plt.xlabel('Goals')
plt.ylabel('Assist')
plt.show()
```

```python
football = pd.read_csv("C:/Users/User/Downloads/dataset - 2020-09-24.csv")

fouls = football['Fouls']
yellow_cards = football['Yellow cards']

plt.figure(figsize=(8,5))
plt.title('Relationship between Fouls vs Yellow Cards')
plt.scatter(fouls,yellow_cards,color="green",alpha= 0.5)
plt.xlabel('Fouls')
plt.ylabel('Yellow cards')
plt.show()
```

```python
players = football.nlargest(8,'Goals')

name = players['Name']
goals = players['Goals']

plt.figure(figsize = (10,8))
plt.pie(goals,labels = name,autopct= '%1.1f%%',startangle=140)
plt.title('Pie chart for Names and Goals')
plt.show()
```
